<a href="https://leetcode.com/problems/minimum-time-to-make-rope-colorful/description/">1578. Minimum Time to Make Rope Colorful</a>

```python
class Solution:
    def minCost(self, colors: str, neededTime: list[int]) -> int:
        total_cost = 0
        n = len(colors)
        last_index = 0
        for i in range(1, n):
            min_time = 0
            if colors[i] == colors[last_index]:
                if neededTime[i] > neededTime[last_index]:
                    min_time = neededTime[last_index]
                    last_index = i
                else:
                    min_time = neededTime[i]
                total_cost += min_time
            else:
                last_index = i
        return total_cost
```

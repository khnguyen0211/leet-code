<a href = "https://leetcode.com/problems/adjacent-increasing-subarrays-detection-ii/description">3350. Adjacent Increasing SubArrays Detection II</a>

```python
class Solution:
    def maxIncreasingSubArrays(self, nums: List[int]) -> int:
        k = 1
        for k in range(k, len(nums) // 2 + 1):
            if not self.hasIncreasingSubArrays(nums, k):
                return k - 1
        return k
    
    def hasIncreasingSubArrays(self, nums: list[int], k: int) -> bool:
        for a in range(len(nums) - k + 1):
            first_arr = nums[a:a + k]
            second_arr = nums[a + k: a + 2*k]
            if(len(first_arr) != len(second_arr)):
                break
            if self.isIncreasing(first_arr) and self.isIncreasing(second_arr):
                return True
        return False
    
    def isIncreasing(self, sub_arr: list[int]) -> bool:
        for i in range(len(sub_arr) - 1):
            if sub_arr[i] >= sub_arr[i + 1]:
                return False
        return True
        
```

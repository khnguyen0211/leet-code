<a href="https://leetcode.com/problems/the-two-sneaky-numbers-of-digitville/description/">3289. The Two Sneaky Numbers of Digitville</a>

```python
from collections import Counter

class Solution:
    def getSneakyNumbers(self, nums: list[int]) -> list[int]:
        result = []
        dict = Counter(nums)
        list_x = list(dict.items())
        for k, v in list_x:
            if v > 1:
                result.append(k)
        return result
```

```python
Solution().getSneakyNumbers([7,1,5,4,3,4,6,0,9,5,8,2])
```

```python
class Solution:
    def getSneakyNumbers(self, nums: list[int]) -> list[int]:
        my_dict = {i : 0 for i in nums}
        result = []
        for i in nums:
            my_dict[i] += 1
            if my_dict[i] == 2:
                result.append(i)
                if len(result) == 2:
                    return result
        return result
```

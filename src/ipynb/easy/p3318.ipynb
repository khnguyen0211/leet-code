<a href="https://leetcode.com/problems/find-x-sum-of-all-k-long-subarrays-i/description/">3318. Find X-Sum of All K-Long Subarrays I</a>

```python
from collections import Counter

class Solution:
    def findXSum(self, nums: list[int], k: int, x: int) -> list[int]:
        x_sum_arr = list(range(len(nums) - k + 1))
        for i in range(len(nums) - k + 1):
            sub_array = nums[i:i + k]
            x_sum_arr[i] = self.sum(sub_array, x)
        return x_sum_arr

    def sum(self, arr: list[int], x: int) -> int:
        my_dict = Counter(arr)
        sorted_dict = dict(sorted(my_dict.items(), key=lambda x: (-x[1], -x[0])))
        top_x = dict(list(sorted_dict.items())[:x])
        total = sum(key * value for key, value in top_x.items())
        return total

Solution().findXSum([1,1,2,2,3,4,2,3], 6, 2)
```

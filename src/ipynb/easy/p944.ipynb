<a href="https://leetcode.com/problems/delete-columns-to-make-sorted/">944. Delete Columns to Make Sorted</a>

```python
class Solution:
    def minDeletionSize(self, strs: list[str]) -> int:
        column_length = len(strs[0])
        counter = 0
        for i in range(column_length):
            for j in range(1, len(strs)):
                if strs[j][i] < strs[j - 1][i]:
                    counter += 1
                    break
        return counter
```

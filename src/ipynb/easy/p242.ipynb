<a href="https://leetcode.com/problems/valid-anagram/description/">242. Valid Anagram</a>

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        dict_s = {}
        for i in s:
            if i in dict_s:
                dict_s[i] += 1
            else:
                dict_s[i] = 1
        for i in t:
            if i in dict_s:
                dict_s[i] -= 1
            else:
                return False
        for _, val in dict_s.items():
            if val != 0:
                return False
        return True
```
